```python
restaurantes = merged_data.copy()
```

```python
restaurantes['Lat&Lon'] = restaurantes['Lat&Lon'].astype(str)


lat_lon_pattern = r'\(([^,]+), ([^)]+)\)'
restaurantes[['lat', 'long']] = restaurantes['Lat&Lon'].str.extract(lat_lon_pattern)
```

```python
# Convierte las nuevas columnas a tipos numéricos si es necesario
restaurantes[['lat', 'long']] = restaurantes[['lat', 'long']].apply(pd.to_numeric, errors='coerce')

# Elimina la columna "Lat&Lon"
restaurantes = restaurantes.drop('Lat&Lon', axis=1)
```

```python
restaurantes.head(15)
```

```python
restaurantes['lat'] = restaurantes['lat'].fillna('No disponible')
restaurantes.loc[restaurantes['ubicacion'] == 'No disponible', 'lat'] = 'No disponible'
```

```python
restaurantes['long'] = restaurantes['long'].fillna('No disponible')
restaurantes.loc[restaurantes['ubicacion'] == 'No disponible', 'long'] = 'No disponible'
```

```python
#IMPORTO EL DATAFRAME RESULTANTE A UN CSV.

restaurantes.to_csv('../datos/rest_clean.csv', index=False, encoding='utf-8')
```
