```python
import pandas as pd
import numpy as np
import geopandas as gpd
import folium
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
#Importo los df completo de cada variable, con las coordenadas, punto y barrio al que pertenecen. 

restaurantes = pd.read_csv('../csv/restaurante_fin.csv')
eventos = pd.read_csv('../csv/eventos_fin.csv')
geo = gpd.read_file('./barrios-madrid.geojson')
```

```python
# Nos guiamos por la mediana y media para señalar los precios por valores: bajos, medios, altos (en cuanto a los eventos seleccionados)

mediana_media = eventos['precio'].median(), eventos['precio'].mean()
mediana_media
```

```python
type('precio')
```

```python

bins = [0, 25, 60, float('inf')]  # límites de las categorías
labels = ['bajo', 'medio', 'alto']  # etiquetas correspondientes a cada categoría

# Crea la nueva columna 'categoria_precio'
eventos['tax'] = pd.cut(eventos['precio'], bins=bins, labels=labels, right=False)
```

Voy a seleccionar aquellas columnas que me son más utiles para llevar a cabo el mapa. Y a través de las cuales puedo hacer una visualización sencilla y clara. 

```python
eventos = eventos[['nombre','local', 'tax', 'barrio','lat', 'long']]
restaurantes = restaurantes[['restaurante','gastronomia', 'precio', 'barrio','lat', 'long']]
```

```python
restaurantes.insert(1,'tipo', 'restaurante')
eventos.insert(1, 'tipo', 'evento')
```

```python
eventos.head()
```

```python
restaurantes.head()
```

```python
# crear mapa
mapa = folium.Map(location=[40.416709, -3.690286], 
                  zoom_start=15, 
                  tiles = 'Stamen Terrain')
```

```python
restaurantes = restaurantes.dropna(subset=['lat', 'long'])
```

```python
eventos = eventos.dropna(subset=['lat', 'long'])
```

```python
len(eventos)
```

```python
len(restaurantes)
```

```python
restaurantes['lat'] = pd.to_numeric(restaurantes['lat'], errors='coerce')
restaurantes['long'] = pd.to_numeric(restaurantes['long'], errors='coerce')
```

```python
for index, row in restaurantes.iterrows():
    popup_txt = folium.Popup(row['restaurante'], row['tipo'])
    folium.Marker([float(row['lat']), float(row['long'])], popup=popup_txt, icon=folium.Icon(color='red')).add_to(mapa)
```

```python
restaurantes.head()
```

```python
eventos['lat'] = pd.to_numeric(eventos['lat'], errors='coerce')
eventos['long'] = pd.to_numeric(eventos['long'], errors='coerce')
```

```python
for index, row in eventos.iterrows():
    popup_txt = folium.Popup(row['nombre'], row['tipo'])
    folium.Marker([float(row['lat']), float(row['long'])], popup=popup_txt, icon=folium.Icon(color='red')).add_to(mapa)
```

```python
eventos.head()
```

```python
# Crear un nuevo DataFrame con valores únicos en la columna 'nombre'
nuevos_eventos = eventos.drop_duplicates(subset='nombre')
```

```python
nuevos_eventos.head()
```

```python
restaurantes.head()
```

```python
nuevos_eventos = nuevos_eventos.rename(columns={'nombre': 'nombre', 'tax': 'precio'})
restaurantes = restaurantes.rename(columns={'restaurante': 'nombre'})
```

```python
mapa_concat = pd.concat([restaurantes, nuevos_eventos], axis = 0)
```

```python
mapa_concat.tail()
```

```python
nuevos_eventos.to_csv('./csv_map/eventos_map.csv', index = False)
restaurantes.to_csv('./csv_map/restaurantes_map.csv', index = False)
mapa_concat.to_csv('./csv_map/madrid_map.csv', index = False)
```
