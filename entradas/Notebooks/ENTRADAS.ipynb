```python
num_duplicates = url.Enlace.duplicated().value_counts()
num_duplicates
```

```python
#Elimino aquellas filas en las que el enlace al evento se repite, ya que entiendo que podría ensuciarnos los datos. 

url.drop_duplicates(subset='Enlace', inplace=True)
```

```python
len(url)
```

```python
# Escribe los nombres de las columnas en minusculas

url.columns = url.columns.str.lower()
```

```python
# Eliminamos esta fila, ya que contiene un pack que no nos interesa y podría darnos problemas.

indice_a_eliminar = 1693
url = url.drop(indice_a_eliminar)

# Para que los índices se reorganicen después de eliminar la fila
url.reset_index(drop=True, inplace=True)
```

```python
# La columna 'fecha' la pasamos a datetime y así podremos hacer los cambios que queremos.
# como tengo año, mes, dia y hora, voy a separarlo todo, de momento. 

url['fecha'] = pd.to_datetime(url['fecha'], utc=True)
```

```python
url['año'] = url['fecha'].dt.year
url['mes'] = url['fecha'].dt.month
url['dia'] = url['fecha'].dt.day
url['hora'] = url['fecha'].dt.time
```

```python
# ELIMINO LA COLUMNA DE FECHA

url.drop('fecha', axis=1, inplace=True)
```

```python
url.head()
```

```python
url_clean = url.copy()
```

```python
# LIMPIO LA COLUMNA 'precio' Y LE ELIMINO EL "desde"

url_clean['precio'] = url_clean['precio'].str.replace('desde', '').str.replace(',', '.').str.strip()

url_clean.head()
```
